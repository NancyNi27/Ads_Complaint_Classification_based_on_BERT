```python
import os
import re
from typing import List, Tuple
from pathlib import Path
import PyPDF2

def extract_file_ids(filenames: List[str]) -> List[Tuple[str, str]]:
    """
    从文件名中提取数字ID。
    返回一个包含(文件名, ID)元组的列表。
    """
    result = []
    seen_ids = set()
    
    for filename in filenames:
        # 使用正则表达式匹配文件名中的数字模式
        # 匹配形如 17_024, 17_091, 20_285 等模式
        match = re.search(r'(\d+[_]\d+|\d+)', filename)
        
        if match:
            # 获取匹配到的数字部分
            id_str = match.group(1)
            # 移除下划线，保证ID格式统一
            id_str = id_str.replace('_', '')
            
            # 如果ID已存在，添加后缀以避免重复
            original_id = id_str
            counter = 1
            while id_str in seen_ids:
                id_str = f"{original_id}_{counter}"
                counter += 1
            
            seen_ids.add(id_str)
            result.append((filename, id_str))
        else:
            print(f"警告: 无法从文件名 '{filename}' 中提取ID")
    
    return result

def convert_pdf_to_txt(pdf_path: str) -> str:
    """
    将PDF文件转换为文本。
    返回提取的文本内容。
    """
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            for page in pdf_reader.pages:
                text += page.extract_text() + "\n"
    except Exception as e:
        print(f"转换PDF文件 '{pdf_path}' 时出错: {str(e)}")
        return ""
    return text

def process_pdf_files(input_dir: str, output_dir: str):
    """
    处理指定目录下的所有PDF文件，将其转换为txt并使用提取的ID重命名。
    """
    # 创建输出目录（如果不存在）
    Path(output_dir).mkdir(parents=True, exist_ok=True)
    
    # 获取所有PDF文件
    pdf_files = [f for f in os.listdir(input_dir) if f.lower().endswith('.pdf')]
    
    # 提取文件ID
    file_ids = extract_file_ids(pdf_files)
    
    # 处理每个文件
    for filename, id_str in file_ids:
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, f"{id_str}.txt")
        
        print(f"处理文件: {filename} -> {id_str}.txt")
        
        # 转换PDF到文本
        text_content = convert_pdf_to_txt(input_path)
        
        # 保存文本文件
        if text_content:
            try:
                with open(output_path, 'w', encoding='utf-8') as f:
                    f.write(text_content)
                print(f"成功转换: {filename}")
            except Exception as e:
                print(f"保存文件 '{output_path}' 时出错: {str(e)}")
        else:
            print(f"警告: 无法从 '{filename}' 提取文本")

if __name__ == "__main__":
    # 设置输入和输出目录
    input_directory = "/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/input/ASA/data2"
    output_directory = "/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/work/output"
    
    # 处理文件
    process_pdf_files(input_directory, output_directory)
```

# 批量处理input/data_n/.pdf

```python
import os
import re
import csv
from typing import List, Tuple
from pathlib import Path
import PyPDF2

def extract_file_ids(filenames: List[str]) -> List[Tuple[str, str]]:
    """
    从文件名中提取数字ID。
    返回一个包含(文件名, ID)元组的列表。
    """
    result = []
    seen_ids = set()
    
    for filename in filenames:
        match = re.search(r'(\d+[_]\d+|\d+)', filename)
        
        if match:
            id_str = match.group(1)
            id_str = id_str.replace('_', '')
            
            original_id = id_str
            counter = 1
            while id_str in seen_ids:
                id_str = f"{original_id}_{counter}"
                counter += 1
            
            seen_ids.add(id_str)
            result.append((filename, id_str))
        else:
            print(f"警告: 无法从文件名 '{filename}' 中提取ID")
    
    return result

def convert_pdf_to_txt(pdf_path: str) -> str:
    """
    将PDF文件转换为文本。
    返回提取的文本内容。
    """
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            for page in pdf_reader.pages:
                text += page.extract_text() + "\n"
    except Exception as e:
        print(f"转换PDF文件 '{pdf_path}' 时出错: {str(e)}")
        return ""
    return text

def process_pdf_files(input_dir: str, output_dir: str, mapping_data: list):
    """
    处理指定目录下的所有PDF文件，将其转换为txt并使用提取的ID重命名。
    同时收集文件名映射信息。
    """
    Path(output_dir).mkdir(parents=True, exist_ok=True)
    
    pdf_files = [f for f in os.listdir(input_dir) if f.lower().endswith('.pdf')]
    
    if not pdf_files:
        print(f"警告: 在目录 '{input_dir}' 中没有找到PDF文件")
        return
    
    file_ids = extract_file_ids(pdf_files)
    
    for filename, id_str in file_ids:
        input_path = os.path.join(input_dir, filename)
        txt_filename = f"{id_str}.txt"
        output_path = os.path.join(output_dir, txt_filename)
        
        print(f"处理文件: {filename} -> {txt_filename}")
        
        text_content = convert_pdf_to_txt(input_path)
        
        if text_content:
            try:
                with open(output_path, 'w', encoding='utf-8') as f:
                    f.write(text_content)
                print(f"成功转换: {filename}")
                # 添加映射信息到列表
                mapping_data.append([filename, txt_filename])
            except Exception as e:
                print(f"保存文件 '{output_path}' 时出错: {str(e)}")
        else:
            print(f"警告: 无法从 '{filename}' 提取文本")

def get_data_folder_range(base_path: str) -> tuple[int, int]:
    """
    自动检测data文件夹的范围。
    返回(最小数字, 最大数字)的元组。
    """
    folders = os.listdir(base_path)
    data_folders = []
    
    pattern = re.compile(r'data(\d+)$')
    
    for folder in folders:
        match = pattern.match(folder)
        if match:
            folder_num = int(match.group(1))
            data_folders.append(folder_num)
    
    if not data_folders:
        raise ValueError(f"在 {base_path} 中没有找到data文件夹")
    
    return min(data_folders), max(data_folders)

def process_multiple_folders(base_input_path: str, base_output_path: str):
    """
    自动检测并处理所有data文件夹中的PDF文件。
    同时生成文件名映射的CSV文件。
    """
    try:
        start_folder, end_folder = get_data_folder_range(base_input_path)
        print(f"检测到data文件夹范围: data{start_folder} 到 data{end_folder}")
    except ValueError as e:
        print(f"错误: {str(e)}")
        return
    
    total_processed = 0
    total_failed = 0
    
    # 用于存储所有文件名映射的列表
    mapping_data = [["PDF文件名", "TXT文件名"]]  # CSV header
    
    for folder_num in range(start_folder, end_folder + 1):
        folder_name = f"data{folder_num}"
        input_dir = os.path.join(base_input_path, folder_name)
        
        if not os.path.exists(input_dir):
            print(f"跳过不存在的文件夹: {input_dir}")
            continue
            
        print(f"\n开始处理文件夹: {folder_name}")
        print("-" * 50)
        
        try:
            process_pdf_files(input_dir, base_output_path, mapping_data)
            total_processed += 1
        except Exception as e:
            print(f"处理文件夹 {folder_name} 时出错: {str(e)}")
            total_failed += 1
    
    # 保存映射信息到CSV文件
    mapping_file = os.path.join(base_output_path, "filename_mapping.csv")
    try:
        with open(mapping_file, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerows(mapping_data)
        print(f"\n文件名映射已保存到: {mapping_file}")
    except Exception as e:
        print(f"保存映射文件时出错: {str(e)}")
            
    print("\n处理完成统计:")
    print(f"成功处理文件夹数: {total_processed}")
    print(f"失败文件夹数: {total_failed}")

if __name__ == "__main__":
    base_input_path = "/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/input/ASA"
    base_output_path = "/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/work/output"
    
    process_multiple_folders(base_input_path, base_output_path)
```

```python
import csv
import re

def count_newer_files(csv_file: str):
    """
    统计TXT_name列中以18或更大数字开头的文件数量
    """
    with open(csv_file, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        
        count = 0
        matched_files = []
        
        for row in reader:
            txt_name = row['TXT_name']
            # 提取txt文件名开头的数字
            match = re.match(r'^(\d+)', txt_name)
            if match:
                number = int(match.group(1))
                if number >= 18:
                    count += 1
                    matched_files.append((txt_name, row['PDF_name']))
        
        print(f"以18或更大数字开头的文件数量: {count}")
        print("\n符合条件的文件列表:")
        print("-" * 50)
        for txt_name, pdf_name in matched_files:
            print(f"TXT: {txt_name}")
            print(f"PDF: {pdf_name}")
            print()

if __name__ == "__main__":
    csv_path = "/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/work/output/filename_mapping.csv"
    count_newer_files(csv_path)
```

### 删除所有空行

```python
import os
from pathlib import Path
from datetime import datetime

def remove_leading_empty_lines(output_dir: str):
    """
    删除所有txt文件开头的空行
    """
    # 记录处理的文件
    changed_files = []
    unchanged_files = []
    error_files = []
    
    # 获取所有txt文件
    txt_files = [f for f in os.listdir(output_dir) if f.endswith('.txt')]
    
    for filename in txt_files:
        file_path = os.path.join(output_dir, filename)
        print(f"处理文件: {filename}")
        
        try:
            # 读取文件内容
            with open(file_path, 'r', encoding='utf-8') as f:
                lines = f.readlines()
            
            # 找到第一个非空行
            first_non_empty = 0
            while first_non_empty < len(lines) and not lines[first_non_empty].strip():
                first_non_empty += 1
            
            # 如果有空行需要删除
            if first_non_empty > 0:
                # 保存修改后的内容
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.writelines(lines[first_non_empty:])
                changed_files.append(filename)
                print(f"已删除 {first_non_empty} 个空行: {filename}")
            else:
                unchanged_files.append(filename)
                print(f"文件开头没有空行: {filename}")
                
        except Exception as e:
            error_files.append((filename, str(e)))
            print(f"处理文件 {filename} 时出错: {str(e)}")
    
    # 生成日志文件
    log_filename = f"empty_lines_removal_log2_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
    log_path = os.path.join(output_dir, log_filename)
    
    with open(log_path, 'w', encoding='utf-8') as f:
        f.write("=== 空行删除操作日志 ===\n")
        f.write(f"执行时间: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n")
        
        f.write("=== 已修改的文件 ===\n")
        f.write(f"总数: {len(changed_files)}\n")
        for file in changed_files:
            f.write(f"- {file}\n")
        
        f.write("\n=== 未修改的文件 ===\n")
        f.write(f"总数: {len(unchanged_files)}\n")
        for file in unchanged_files:
            f.write(f"- {file}\n")
        
        if error_files:
            f.write("\n=== 处理出错的文件 ===\n")
            f.write(f"总数: {len(error_files)}\n")
            for file, error in error_files:
                f.write(f"- {file}: {error}\n")
    
    print(f"\n处理完成:")
    print(f"- 已修改: {len(changed_files)} 个文件")
    print(f"- 未修改: {len(unchanged_files)} 个文件")
    print(f"- 出错: {len(error_files)} 个文件")
    print(f"详细日志已保存到: {log_filename}")

if __name__ == "__main__":
    output_dir = "/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/work/output"
    remove_leading_empty_lines(output_dir)
```

### 删除所有固定无意义的开头

```python
import os
from pathlib import Path
from datetime import datetime

def remove_standard_header(output_dir: str):
    """
    删除所有txt文件中的标准四行开头文本
    """
    # 要删除的四行标准文本
    header_lines = [
        "Thank you for contacting the A dvertising Standards Authority  with your concerns about this ",
        "advertisement. We value your feedback. In accordance with our Rules, we can only accept",
        "complaints where there is a likely breach of one or more of the  Advertising Standards",
        "Authority Codes ."
    ]
    
    # 记录处理的文件
    changed_files = []
    unchanged_files = []
    error_files = []
    
    # 获取所有txt文件
    txt_files = [f for f in os.listdir(output_dir) if f.endswith('.txt')]
    total_files = len(txt_files)
    
    print(f"开始处理 {total_files} 个文件...")
    
    for filename in txt_files:
        file_path = os.path.join(output_dir, filename)
        print(f"处理文件: {filename}")
        
        try:
            # 读取文件内容
            with open(file_path, 'r', encoding='utf-8') as f:
                lines = f.readlines()
            
            # 检查前四行是否匹配（考虑到可能有空行，需要找到实际的文本行）
            content_lines = [line.strip() for line in lines if line.strip()]
            
            if len(content_lines) >= 4 and all(content_lines[i].strip() == header_lines[i].strip() for i in range(4)):
                # 删除这四行，保留其他内容
                new_content = '\n'.join(content_lines[4:])
                
                # 保存修改后的内容
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write(new_content)
                changed_files.append(filename)
                print(f"✓ 成功删除标准开头: {filename}")
            else:
                unchanged_files.append(filename)
                print(f"- 文件开头不匹配标准格式: {filename}")
                
        except Exception as e:
            error_files.append((filename, str(e)))
            print(f"× 处理文件 {filename} 时出错: {str(e)}")
    
    # 生成日志文件
    log_filename = f"header_removal_log_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
    log_path = os.path.join(output_dir, log_filename)
    
    with open(log_path, 'w', encoding='utf-8') as f:
        f.write("=== 标准开头删除操作日志 ===\n")
        f.write(f"执行时间: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"总文件数: {total_files}\n\n")
        
        f.write("要删除的标准开头内容：\n")
        for i, line in enumerate(header_lines, 1):
            f.write(f"第{i}行: {line}\n")
        f.write("\n")
        
        f.write("=== 已修改的文件 ===\n")
        f.write(f"总数: {len(changed_files)}\n")
        for file in changed_files:
            f.write(f"- {file}\n")
        
        f.write("\n=== 未修改的文件 ===\n")
        f.write(f"总数: {len(unchanged_files)}\n")
        for file in unchanged_files:
            f.write(f"- {file}\n")
        
        if error_files:
            f.write("\n=== 处理出错的文件 ===\n")
            f.write(f"总数: {len(error_files)}\n")
            for file, error in error_files:
                f.write(f"- {file}: {error}\n")
    
    print(f"\n处理完成:")
    print(f"- 总文件数: {total_files}")
    print(f"- 已修改: {len(changed_files)} 个文件")
    print(f"- 未修改: {len(unchanged_files)} 个文件")
    print(f"- 出错: {len(error_files)} 个文件")
    print(f"详细日志已保存到: {log_filename}")

if __name__ == "__main__":
    output_dir = "/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/work/output"
    remove_standard_header(output_dir)
```

```python
import os
import datetime

def check_file_format(directory):
    # 获取当前时间戳用于日志文件名
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    log_filename = f"format_check_log_{timestamp}.txt"
    
    # 预期的每行开头内容
    expected_starts = [
        "COMPLAINT NUMBER",
        "ADVERTISER",
        "ADVERTISEMENT",
        "DATE OF DECISION",
        "OUTCOME"
    ]
    
    # 创建或打开日志文件
    with open(log_filename, 'w', encoding='utf-8') as log_file:
        log_file.write(f"Format Check Log - {datetime.datetime.now()}\n")
        log_file.write("Checking files in directory: work/output/\n")
        log_file.write("-" * 50 + "\n\n")
        
        # 遍历目录中的所有.txt文件
        for filename in os.listdir(directory):
            if filename.endswith('.txt'):
                file_path = os.path.join(directory, filename)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        # 读取前5行
                        lines = [line.strip() for line in f.readlines()[:5]]
                        
                    # 检查文件是否有至少5行
                    if len(lines) < 5:
                        log_file.write(f"File: {filename}\n")
                        log_file.write(f"Error: File has less than 5 lines (only {len(lines)} lines)\n\n")
                        continue
                    
                    # 检查每行的开头
                    for i, (line, expected) in enumerate(zip(lines, expected_starts)):
                        if not line.startswith(expected):
                            log_file.write(f"File: {filename}\n")
                            log_file.write(f"Error: Line {i+1} does not start with '{expected}'\n")
                            log_file.write(f"Found: {line[:50]}{'...' if len(line) > 50 else ''}\n\n")
                            break
                            
                except Exception as e:
                    log_file.write(f"File: {filename}\n")
                    log_file.write(f"Error: Failed to process file - {str(e)}\n\n")
        
        log_file.write("\nCheck completed.")
    
    return log_filename

# 执行检查
result_log = check_file_format("/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/work/output")
print(f"检查完成，日志文件已生成: {result_log}")
```

```python
import os
import datetime

def check_file_format(directory):
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    log_filename = f"format_check_log_{timestamp}.txt"
    
    # 定义两种可接受的格式
    format_1 = [
        "COMPLAINT NUMBER",
        "ADVERTISER",
        "ADVERTISEMENT",
        "DATE OF DECISION",
        "OUTCOME"
    ]
    
    format_2 = [
        "COMPLAINT NUMBER",
        "COMPLAINANT",
        "ADVERTISER",
        "ADVERTISEME",
        "DATE OF MEETING",
        "OUTCOME"
    ]
    
    valid_starts = ['18', '19', '20', '22', '23', '24']
    
    with open(log_filename, 'w', encoding='utf-8') as log_file:
        log_file.write(f"Format Check Log - {datetime.datetime.now()}\n")
        log_file.write("Checking files in directory: work/output/\n")
        log_file.write("Only checking files starting with: " + ", ".join(valid_starts) + "\n")
        log_file.write("-" * 50 + "\n\n")
        
        for filename in os.listdir(directory):
            if filename.endswith('.txt') and any(filename.startswith(start) for start in valid_starts):
                file_path = os.path.join(directory, filename)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        lines = [line.strip() for line in f.readlines()[:6]]  # 读取前6行以适应更长的格式
                    
                    # 检查是否符合任一格式
                    format1_valid = True
                    format2_valid = True
                    
                    # 检查格式1
                    if len(lines) >= 5:
                        for i, expected in enumerate(format_1):
                            if not lines[i].startswith(expected):
                                format1_valid = False
                                break
                    else:
                        format1_valid = False
                    
                    # 检查格式2
                    if len(lines) >= 6:
                        for i, expected in enumerate(format_2):
                            if not lines[i].startswith(expected):
                                format2_valid = False
                                break
                    else:
                        format2_valid = False
                    
                    # 如果两种格式都不符合，记录到日志
                    if not format1_valid and not format2_valid:
                        log_file.write(f"File: {filename}\n")
                        log_file.write("Error: File does not match either format\n")
                        log_file.write("Found lines:\n")
                        for i, line in enumerate(lines):
                            log_file.write(f"Line {i+1}: {line[:50]}{'...' if len(line) > 50 else ''}\n")
                        log_file.write("\n")
                            
                except Exception as e:
                    log_file.write(f"File: {filename}\n")
                    log_file.write(f"Error: Failed to process file - {str(e)}\n\n")
        
        log_file.write("\nCheck completed.")
    
    return log_filename

# 执行检查
result_log = check_file_format("/Users/niwenyu/Desktop/OCR_PDF_EXTRACT/work/output")
print(f"检查完成，日志文件已生成: {result_log}")
```
